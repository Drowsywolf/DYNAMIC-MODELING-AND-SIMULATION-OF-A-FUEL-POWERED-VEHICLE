### 速度数据
0~6003都为0，实测数据为6004~68042，共62039个数据，每秒约20个数据

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
path = r'Tiguan-Speed-Drivecycle.csv'
datadf = pd.read_csv(path)
data = datadf.values
```

```python
plt.figure(figsize=(10, 5))

plt.plot(data[:,0],data[:,1])
plt.xlabel(r'时间[s]', font="MicroSoft YaHei")
plt.ylabel(r"速度[km/h]", font="MicroSoft YaHei")
```
